## Checking the OLS assumptions

Price, Year, Mileage and Engive V are the most worrying variables because they are continuous. Discrete variables are less strict will be computed as dummies.

In order to ensure OLS assumptions, continuous variables follow some presumptions. 
The first assumption is linearity:  first and foremost, we have to check whether the relationships between inputs and the output follows a linear pattern. If they aren't we adjust them accordingly.

```python
f, (ax1,ax2,ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')
plt.show()
```

```python
'''All relationships between price and input variables don't seem
to follow a linear pattern. In fact, the distribution of price is more
alike an exponential distribution than a normal one. What are the implications?
Usually, in this situation the best solution is logarithmize the variables.'''

#numpy has a log function
log_price = np.log(data_cleaned['Price']) 
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
#Maybe with the logarithmization of price we can actually get a pattern that is more resemblant of a linear regression.
#First, we test log_price against our inputs

f, (ax1,ax2,ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')
plt.show()

#We can see that there is much more linear friendly pattern.
```

Given that it works, we can remove altogether the price variable.

```python
data_cleaned = data_cleaned.drop(['Price'],axis = 1)
```

The second assumption is no Endogeneity. In essence, what this means is that we have to ensure that there is no correlation between the inputs and the unobserved features of the model (the error).

The third assumption is Normality and Homocedasticity:

We made our efforts to ensure normality of the model. Plus given our sample size, we can ensure normality by applying the Central Limit Theorem which foresees that any distribution, as its sample size tends to infinity, tends to follow a normal distribution.

By including the intercept in the model, we ensure that there is 0 mean.

We handled the homocedasticity by logarithmizing our output.

Finally, we need to ensure that there is no auto-correlation between our sample's input observations. Considering that our sample comes from a second-har car shop and we are not looking at a time-series or panel data. It is hard to argue that there is auto-correlation between different cars.

### Dealing with multicollinearity

```python
#first, we check for multicollinearity
data_cleaned.columns.values
```

```python
#There is a discussion to be had when it comes to the relationship between Year and Mileage (The more time passes, the more likely it is that a car has a larger mileage)
#There is no sklearn for this. It is wise to import VIF from statsmodels

from statsmodels.stats.outliers_influence import variance_inflation_factor
#we check for multicolinearity between our continuous input variables
variables = data_cleaned[['Mileage','Year','EngineV']]
#create a dataframe with 2 columns, one for each variable and the other for the VIF of each variable
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values,i) for i in range (variables.shape[1])]
vif['Features'] = variables.columns
```

```python
vif #VIF is an unusual variable. For the sake of this exercise, we'll exclude any value with a Covariance Inflation Factor of 10+. The literature, however, is not unanimous as a substancial amount of researchers considers VIFs above 5 to be unacceptable.
```

```python
#We'll remove year due to its high multicolinearity value:
data_no_multicollinearity = data_cleaned.drop(['Year'], axis = 1)
```

So far we've dealt mainly with the continuous variables and let discrete/categorical ones. On this section, we are going to include dumy variables and make adjustments as necessary.

```python
#pandas has a get dummies method, let's use it:

data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies.head()
'''We can see that dummies were created concerning the car Brand and its Engine type'''
```

### Rearrange

```python
data_with_dummies.columns.values
```

```python
#copy-paste the previous and place the dependent variable in the first place
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol']
```

## Standardization and Regression

Ok, we made a lot of processing to our data up to this point. We started by removing blank values, determining which variables could be of potential interest or not, checking whether the OLS assumptions actually hold water and finally rearrange all data accordingly.

Now is the time to to follow up on all of this data treatment and actually compute and plot the data and derive conclusions and potentially make predicitions for possible new data that arises.

```python
#we know our target is price (logarithmized in order to fulfill OLS assumption of linearity)
targets = data_pre_processed['log_price']
#The remaining variables are inputs
inputs = data_pre_processed.drop(['log_price'], axis = 1)
```

## Data Scaling

We don't want to fall in the trap of wrongly deriving conclusions from coefficients. So, in order to correctly ascertain each input's impact on the determination of price, we'll scale the data. As done previously we'll use the sklearn StandardScaler. 

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#we fit the scaler measurements to our sample. We do that by the fit method
scaler.fit(inputs)
```

```python
#we now convert our inputs to scaled
inputs_scaled = scaler.transform(inputs)
#Thus, it returns an array of our scaled inputs
inputs_scaled
```

It is not a good idea to scale dummy variables. This happens because they lose their dummy meaning.

However, in ML, they keep their predictive power.

### Training and Testing

We have to be wary of possible overfitting of the model. Overfitting, in essence is a process by which the ML model follows the sample data to an excessive degree. In theory this should be a good thing. However, overfitting means that the model adjusted so well to the data that, actually, it fits the data (and its outliers) so much, that patterns and trends become overshadowed.

In practical means, this would mean that the model would be an unreliable predictor. It is hard to spot so, a method usually used is to actually split the data into testing data and training data. The model will be trained with a portion of the data. Afterwards, the remaining portion is used to test the data.

If a model is overfitted, the predictions of the test would have low accuracy. This would mean that, although the model adjusted particularly well to the training data, it failed to grasp the underlying logic behind the data distribution.

```python
from sklearn.model_selection import train_test_split
#we assing to each test and train variables. The method splits them automatically
#test size is 0.2 (20% of the data will be used to test, a random state allows us to make sure that the data distribution follows a particular pattern)
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 42)
```

## Create a Regression

```python
#create regression with the training data. This will be a log-lin regression.
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#let's plot the predicted values of the regression to check the result
#we start by creating y_hat = the prediction of our input training data and check how it resembles the actual y_train values
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Output Training Data', fontsize = 20)
plt.ylabel('Predicitions of output of the model', fontsize = 20)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()

#In general, we can see that the model tends to place our predictions near the real output values. The fit is not perfect, but accpetable.
```

```python
#another target check is the residual plot - the difference between targets and target predictions
# in order to follow our assumptions the errors must have to be normally distributed with a mean of 0.
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF')
#in general, the model seems to fit a normal distribution. We should be wary, however that there is a longer tail on the negative side.
#There might be a tendency for the model to overshoot the price. Which means predict by excess.
```

```python
#let's see the Rsquared
reg.score(x_train, y_train)
#The model is responsible for explaining 65% of the variability of the data
```

Also important to note is the interpretation of dummies: When one dummy is one for brand, all others have the value of 0. 
When all dummies for brand are 0, the car is an Audi (the removed benchmark case):

If a coefficient of a brand is negative (say Mitsubishi's or Toyota's) that means that, all else equal, these cars are cheaper than an Audi.
If the coefficient is positive ( Mercedes) than that means that, all else equal, a Mercedes Benz is more expensive than an Audi.

Our dummy benchmarks are:

Body - CrossOver

Engine Type - Diesel

We should be wary of this. Any dummy coefficient is relative to these benchmarks.

## Testing

So, up until now we only used our training data - i.e. the data from which the coefficients were derived. By plotting the data we were able to rule out underfitting - In general the real targets were close to the predicitions of the training inputs. 

Now is the time we give to our model a new set of data that the model has not seen yet. If the predictions fall close to reality, this means that our model understood the underlying logic.


```python
#we start by creating a variable with the predictions of our testing data:

y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test)
plt.xlabel('Test Data Real Values', fontsize = 20)
plt.ylabel('Predicitions of the model\'s Test Inputs', fontsize = 20)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

In general, we can see that, for this set, the model was more accurate in determining higher prices (the data fits a 45 line and is much more condensed for higher prices than for lower ones). We can see use the alpha method to see where there is higher concentration of values.

```python
plt.scatter(y_test,y_hat_test, alpha = 0.2)
plt.xlabel('Test Data Real Values', fontsize = 20)
plt.ylabel('Predicitions of the model\'s Test Inputs', fontsize = 20)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()

#after seeing the concentration, we conclude that, in fact, the higher concentration is on the 45 line.
```

Well, still, we do not a real grasp of how similar the predicted prices are to the real prices. In order to adress that we can create a dataframe with the real prices for the vehicles and the prices predicted by the model.

```python
#we use np.exp because we want the prices, not the logarithm of prices
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ['Prediction'])
df_pf['Targets'] = np.exp(y_test)
df_pf
```

For some reason, we have a lot of missing values in y_test. Why is that?

```python
y_test
#when we split the data, the indexes were preserved. However, the prediction data did not have the same indexes.
#This led to, when computing the for prices, that pandas tried to match indices.
#In order to solve this, we have to remove the indices.
```

```python
y_test = y_test.reset_index(drop = True)
```

```python
#we can try now to check whether the problem was solved of df_pf
df_pf['Targets'] = np.exp(y_test)
df_pf
```

```python
#following the OLS assumptions, it is important to keep in mind what the residual is
# the lower the difference between the real price and its predicition, the better the model's explanatory power
df_pf['Residual'] = df_pf['Targets'] - df_pf['Prediction']
df_pf
```

```python
df_pf['%Difference'] = np.absolute(df_pf['Residual']/df_pf['Targets']*100)
df_pf
```

```python
#We can use a description table to summarize all this
df_pf.describe()
```

The minimum difference was 0.08% - For some cars, the model was a spot on predictor.

The maximum difference however is pretty off the mark (almost 1700%). For most cars, the model was a fairly good prediction (at the 75% percentile the difference between the real price and the predicted price was 50%). It would be interesting to see the data to make a more informed opinion.

```python
#allow us to see the entire data
pd.options.display.max_rows = 999
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df_pf.sort_values(by=['%Difference'])
```
